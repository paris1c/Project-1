```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
owid_covid_data_csv_path = "/Users/victoriasandoval/Project-1/Data/owid-covid-data.csv"
covid_data_df = pd.read_csv(owid_covid_data_csv_path)
covid_data_df.head(2)
```

```python
covid_data_df["date"] = pd.to_datetime(covid_data_df["date"])
```

```python
covid_data_thru_march_world = covid_data_df.loc[(covid_data_df["date"] >= "2020-01-01") & 
                                    (covid_data_df["date"] < "2022-03-31") & 
                                    (covid_data_df["location"] == "World")]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data=covid_data_thru_march_world, x='date', y='new_deaths')
plt.xlabel('Time Period')
plt.ylabel('New Cases')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data=covid_data_thru_march_world, x='date', y='new_cases')
plt.xlabel('Time Period')
plt.ylabel('New Cases')
plt.show()
```

```python
covid_data_2022_03_31 = covid_data_df.loc[(covid_data_df["date"] == "2022-03-31")]
```

```python
covid_data_continents_2022_03_31 = covid_data_2022_03_31.loc[(covid_data_2022_03_31["continent"] == "Asia") | 
                                            (covid_data_2022_03_31["continent"] == "Europe") | 
                                            (covid_data_2022_03_31["continent"] == "Africa") | 
                                            (covid_data_2022_03_31["continent"] == "North America") | 
                                            (covid_data_2022_03_31["continent"] == "South America") | 
                                            (covid_data_2022_03_31["continent"] == "Oceania")]
```

```python
top_10_by_cases = covid_data_continents_2022_03_31.nlargest(10, 'total_cases')
top_10_by_cases
```

```python
plt.figure(figsize=(14,6))
plt.bar("location", "population", data=top_10_by_cases)
plt.bar("location", "total_cases", data=top_10_by_cases)

plt.xlabel("Country", size=15)
plt.ylabel("COVID-19 Cases", size=15)
#plt.ylim( [ 0, 100000000 ] )  
plt.title("Top 10 Countries by COVID-19 Cases (effective 03.31.2022)", size=18)
#plt.savefig("bar_plot_matplotlib_Python.png")
```
